```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"Crime_Data_LAPD.csv")
df.head(10)
```

```python
duplicates = df.duplicated()
duplicates.any()
```

```python
df.info()
```

```python
sorted(df["Victim_age"].unique())
```

```python
#checking for victim age that are < 0
df[df["Victim_age"]<0]
```

```python
#convert the negative age to position in victim column and assign it back to df[victim_age]
df["Victim_age"] = df["Victim_age"].abs()
```

```python
df[df["Victim_age"]==0]
```

```python
#removing the all rows with victim age = 0 at thier indexes and using inplace true meaning it 
# performs the operation on the orignal df without it, it will return a new df and we wld ve reasign 

df.drop(df[df["Victim_age"]==0].index, inplace=True)
```

```python
df.info()
```

```python
#renaming some of the column header 
df.rename(columns={"Date_Reported" : "Date_reported", "Crime_Code" : "Crime_code"}, inplace=True)
```

```python
#to see more of the dataset
pd.set_option("display.max_rows", 100)
```

```python
df.head(100)
```

```python
#check for null values in the datasets
df.isnull().sum()
```

```python
#df['LOCATION'].astype(str)
df[ df['LOCATION'] == 'STOCKER ST']

```

```python
#to delete the rows with null values we can do this below
df.dropna(subset=["Victim_sex", "Victim_descent", "Premis"], inplace=True)
```

```python
#work on the Time_occured column , we convert it to str with astype 4 for to manipulate it
# becos we cant manipulate it as integer & zfill(4)will make it 4 character by adding 0s to the front)

df["Time_occured"]= df["Time_occured"].astype(str).str.zfill(4)

#this is use to make like 18:00 by doing this

#df["Time_occured"] = df["Time_occured"].str[:2]+":"+df["Time_occured"].str[2:]
```

```python
#change the date to date time and using error coerce to make any value inside as not a datetime

df["Date_reported"] = pd.to_datetime(df["Date_reported"],errors="coerce")
df["Date_occured"] = pd.to_datetime(df["Date_occured"], errors="coerce")
```

```python
#Adding new column for Days to report = Date_Reported - Date_Occured and it ll be in days

df["Days_to_report"]= (df["Date_reported"] - df["Date_occured"]).dt.days
```

```python
#Victim analysis this help to know typical victim age so officers can be focus on certain age
# so we count victim by age..we can use histplot

plt.figure(figsize=(15,6))
sns.histplot(x="Victim_age", data=df, bins=10)

```

```python
plt.figure(figsize=(15,6))
sns.countplot(x="Victim_sex", data=df)
```

```python
plt.figure(figsize=(15,6))
sns.countplot(x="Victim_descent", data=df)
```

```python
#Type of crime committed and weapon used analysis

df["Crime_code"].nunique()
```

```python
df["Crime_code"].value_counts()
```

```python
# get the top 10 crimes and visualized it. reset_index convert the series to dataframe, nlargest 4 largest

crime_counts = df["Crime_code"].value_counts().nlargest(10).reset_index()
#naming the column
crime_counts.columns = ["Crime_code", "counts"]
crime_counts
sns.barplot(x="counts", y="Crime_code", data=crime_counts)
# plt.xticks(rotation=90) used to flip the column names
```

```python
# top 10 Weapon used in crime analysis
weapon_counts = df["Weapon"].value_counts().nlargest(10).reset_index()
#naming the column
weapon_counts.columns = ["Weapon", "counts"]
weapon_counts
sns.barplot(x="counts", y="Weapon", data=weapon_counts)
```

```python
#crime by weekdays lets create new column to get week days and get week names

df["Weekday"] = df["Date_occured"].dt.day_name()

# this group the weekday and count the size then reset index to convert from series 
# to datafram and name d column crime copunt

crime_count_per_weekday = df.groupby("Weekday").size().reset_index(name="Crime count")
crime_count_per_weekday
```

```python
#the 1st and 2nd syntax is used to order the weekday in the chart
day_order =["Monday","Tuesday","Wednesday","Thursday", "Friday", "Saturday", "Sunday"]
crime_count_per_weekday["Weekday"] = pd.Categorical(crime_count_per_weekday["Weekday"], categories=day_order,ordered=True)
plt.figure(figsize=(14,6))
sns.barplot(x="Weekday", y="Crime count", data=crime_count_per_weekday)
```
